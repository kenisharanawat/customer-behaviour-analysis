```python
import pandas as pd

data = pd.DataFrame({
    "Customer": ["A","B","C","D"],
    "Purchases":[5,3,8,2],
    "Amount":[500,300,900,150]
})

# total spending
total_spending = data["Amount"].sum()
print("Total Spending:", total_spending)

# highest purchasing customer
top_customer = data.sort_values(by="Purchases", ascending=False)
print(top_customer)
```
